```python
pattern = "g"
def replace_1(all_identity,pattern):
    reg = re.compile(r"[0-9]*{}".format(pattern))
    list_a = list(map(lambda x: reg.findall(x),all_identity))
    list_b = []
    for i in range(len(all_identity)):
        text_ = all_identity[i]
        replace_ = list_a[i]
        for rep in replace_:
            text_ = text_.replace(rep,"")
        list_b.append(text_)
    return list_b
```

```python
def find_1(all_identity,find):
    reg = re.compile(find)
    list_a = list(map(lambda x: reg.findall(x),all_identity))
    list_b = [i for i in list_a if len(i) != 0]

    return list_b
```
